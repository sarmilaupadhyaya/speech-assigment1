```python
import pandas as pd
import glob
from collections import defaultdict
import numpy as np
import math
```

```python
files_path = "ema-std/mngu0_s1_txt_1.1.0_subset/"


def filter_phoneme(start, end, texts):
    """
    get the middle frame of a phoneme
    """
    
    middle = (start +end)/2
    print("middle",middle)
    
    idx = texts['time'].sub(middle).abs().idxmin()
    #added double [[]] for one row DataFrame
    texts = texts.loc[[idx]]
    
    return texts
    
```

```python
phoneme_metrics = defaultdict(list)
## get all the file 
lab_files = glob.glob(files_path+'*.lab')
text_files = glob.glob(files_path+'*.txt')

for each_text in text_files:
    
    each_lab = ".".join(each_text.split(".")[:-1]) + ".lab"
    labs = open(each_lab, "r").readlines()[5:]
    labs = [[x.strip() for x in line.split("\t")] for line in labs]
    df_lab = pd.DataFrame(labs, columns=["time","id","phoneme"])
    df_lab["id"]= df_lab.id.apply(lambda x: float(x.strip().split(" ")[0]))
    df_lab = df_lab[["id", "phoneme"]]
    df_lab.columns = ["time","phoneme"]
    
    ## for each file get phoneme position
    ## df for file txt with time
    lines = open(each_text, "r").readlines()
    lines = [[float(x) for x in line.split(" ")] for line in lines]
    df_text = pd.DataFrame(lines, columns = ["T1x", "T1y", "T2x", "T2y", "T3x", "T3y", "LIx", "LIy", "ULx", "ULy", "LLx", "LLy"])
    df_text["time"] = pd.Series(x/1000.0 for x in list(range(0, (len(df_text)+1)*5,5)))
   
    
    for index, row in df_lab.iterrows():
        if index >0:
            
            result = filter_phoneme(df_lab.iloc[index-1]["time"], row["time"], df_text)
            phoneme_metrics[row["phoneme"]].append(result[["ULx", "ULy", "LLx", "LLy"]].values.flatten().tolist())
            
   
```

```python
phoneme_metrics_summary = defaultdict()
print(len(phoneme_metrics))
list_metrics = []
for k,v in phoneme_metrics.items():
    each = dict()
    each["phoneme"] = k
    each["number of occurence"] = len(v)
    total_dist = []
    for e in v:
        dist = math.sqrt( (e[0] - e[2])**2 + (e[1] - e[3])**2 )
        total_dist.append(dist)
    
    each["min"] = round(min(total_dist),2)
    each["max"] = round(max(total_dist),2)
    each["average"] = round(sum(total_dist)/len(total_dist),2)
    each["std"] = round(np.std(total_dist),3)
    
    list_metrics.append(each)
    
    phoneme_metrics_summary[k] = each
```

```python
print("phonemes","\t","number of occurence","\t\t","min","\t\t","max","\t\t","average","\t","std")
for p, dic in phoneme_metrics_summary.items():
    print(p+"\t\t"+str(round(dic["number of occurence"],2))+"\t\t\t\t"+str(round(dic["min"],2))+"\t\t"+str(round(dic["max"],2))+"\t\t"+str(round(dic["average"],2))+"\t\t"+str(round(dic["std"],2)))
```

```python
list_metrics


df = pd.DataFrame(list_metrics)
df = df.sort_values(by="average")
```

```python
df.to_csv("result.csv", index=False)
```
